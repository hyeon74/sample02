```python
from option.BSM import *
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

mkt_prc = 0.31
option_type = 1
#S = 268.55      #기초자산 가격
K = 275.0       #행사가
#T = 9           #만기일
r = 0.01        #이자율
q = 0.00        #배당율

#Prarmeters
K = 100
sigma = 0.25

#Variables
T = np.linspace(0, 1, 100)
S = np.linspace(0, 200, 100)
T, S = np.meshgrid(T, S)

def option_value(S, K, T, r, sigma, option_type):
        d1 = (np.log(S / K) + (r + 0.5 * sigma **2) * T) / (sigma * np.sqrt(T))
        d2 = d1 - sigma * np.sqrt(T)

        if option_type == 'call':
            V = S * stat.norm.cdf(d1) - K * np.exp(-r * T) * stat.norm.cdf(d2)
        else:
            V = K * np.exp(-r * T) * stat.norm.cdf(-d2) - S * stat.norm.cdf(-d1)

        return V





 


```

```python
call_value = option_value(S, K, T, r, sigma, 'call')

trace = go.Surface(x=T, y=S, z=call_value)
data = [trace]
layout = go.Layout(title='call option', scene={'xaxis':{'title':'AA'}, 'yaxis':{'title':'BB' }})
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
